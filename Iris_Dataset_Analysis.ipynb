## 5. Use matplotlib library to plot this data . Give description about these plots.

### Pie Plot
### This pie chart displays Count of each speices in percentage format.

```python
spieces_count = data['spieces'].value_counts()
plt.title('Spieces Count')
spieces_count.plot.pie(y = None,autopct='%1.2f%%',figsize=(10,8),labeldistance=0.6,pctdistance=0.4)
plt.show()
```

### Conclusion:
### This pie chart depicts even distribution of each speices, from that we can deduce that the count of each species is equal.

### Histogram Plot
### Histogram plot is used to display amount of each attributes.

### Sepal Length Count

```python
plt.title('Sepal Length Count')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.hist(sepal_length_array,edgecolor='black', linewidth=1.2)
```

### Conclusion: The above histogram depicts the frequency of sepal length in the interval 5.5 to 5.75 and 6.0 to 6.5 have higher count.Whereas in the interval 7.0 to 7.5 have the least count.

### Sepal Width Count

```python
plt.title('Sepal Width Count')
plt.xlabel('Sepal Width')
plt.ylabel('Count')
plt.hist(sepal_width_array,edgecolor='black', linewidth=1.2)
```

### Conclusion: The above histogram depicts the frequency of sepal width in the interval 3.0 to 3.5 have higher count.Whereas in the interval 4.0 to 4.5 have the least count.

### Petal Length Count

```python
plt.title('Petal Length Count')
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.hist(petal_length_array,edgecolor='black', linewidth=1.2)
```

### Conclusion: The above histogram depicts the frequency of petal length in the interval 1.0 to 1.5 have higher count.Whereas there are no petal length values in the interval 2.0 to 3.0.

### Petal Width Count

```python
plt.title('Petal Width Count')
plt.xlabel('Petal Width')
plt.ylabel('Count')
plt.hist(petal_width_array,edgecolor='black', linewidth=1.2)
```

### Conclusion: The above histogram depicts the frequency of petal width in the interval 0.1 to 0.25 have higher count.Whereas in the interval 0.5 to 0.75 have the least count.

### Species VS Petal

```python
plt.figure(figsize=(12,9))
plt.subplot(1, 2, 1)
plt.bar(spieces_array, petal_length_array,
        width = 0.4)
plt.xlabel("Spieces")
plt.ylabel("Petal Length")
plt.title("Spieces vs Petal Length")

plt.subplot(1, 2, 2)
plt.bar(spieces_array, petal_width_array,
        width = 0.4)
plt.xlabel("Spieces")
plt.ylabel("Petal Width")
plt.title("Spieces vs Petal Width")

plt.suptitle("Species Vs Petal")
plt.show()
```

### Conclusion: The above subplot depicts two graphs which are Species Vs Petal Length and Species Vs Petal Width.As we can see the species "Iris-Virginica" has highest petal length and width both, whereas species "Iris-Setosa"  has lowest petal length and width.

### Species VS Sepal

```python
plt.figure(figsize=(12,9))
plt.subplot(1, 2, 1)
plt.bar(spieces_array, sepal_length_array,
        width = 0.4)
plt.xlabel("Spieces")
plt.ylabel("Sepal Length")
plt.title("Spieces vs Sepal Length")

plt.subplot(1, 2, 2)
plt.bar(spieces_array, sepal_width_array,
        width = 0.4)
plt.xlabel("Spieces")
plt.ylabel("Sepal Width")
plt.title("Spieces vs Sepal Width")

plt.suptitle("Species Vs Sepal")
plt.show()
```

### Conclusion: The above subplot depicts two graphs which are Species Vs Sepal Length and Species Vs Sepal Width.As we can see the species "Iris-Virginica" has the highest sepal length and "Iris-Setosa" has the lowest length, whereas "Iris-Setosa" has highest sepal width and "Iris-Versicolor" has the lowest width.
